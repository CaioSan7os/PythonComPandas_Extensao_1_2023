```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import plotly.express as px
import plotly
#What social platform has influenced your online shopping most?
#qual rede social mais te influencia na hora de comprar online?
                                 #====Social Midia Influence=====#
smi_df = pd.read_csv('WhatsgoodlyData-6.csv')
```

```python
display(smi_df)
```

```python
smi_students_df = smi_df.loc[smi_df['Segment Type'] == 'University']
```

```python
fig = px.pie(smi_students_df,
             values = smi_students_df['Count'],
             names = smi_students_df['Answer'],
             title = 'Rede social mais influente nas compras onlines entre Estudantes')
```

```python
# update traces
fig.update_traces(
    textposition = 'inside',
    textinfo = 'percent+label'
)
fig.update_layout(
    title_font_size = 30,
)
```
